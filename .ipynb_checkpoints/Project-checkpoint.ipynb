```python
df1_6_x = df1.copy()
df1_6_x.drop(["y"], axis = 1, inplace = True)
X = df1_6_x.to_numpy()
X
```

```python
df1_6_y = df1.copy()
df1_6_y.drop(['x1', 'x2', 'x3', 'x4', 'x5','x6','x7','x8','x9','x10','x11','x12','x13'], axis = 1, inplace = True)
df1_6_y

def flatten(t):
    return [item for sublist in t for item in sublist]

y = flatten(df1_6_y.to_numpy())
y
```

```python
#6-7
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model, metrics
 
reg = linear_model.LinearRegression()
 
# train the model using the training sets
reg.fit(X, y)
 
# regression coefficients
b_i = reg.coef_
print('Coefficients: ', b_i)
 
## setting plot style
plt.style.use('fivethirtyeight')
 
## plotting residual errors in training data
plt.scatter(reg.predict(X), reg.predict(X) - y,
            color = "green", s = 10, label = 'data')
 
## plotting residual errors in test data
#plt.scatter(reg.predict(X_test), reg.predict(X_test) - y_test,
#           color = "blue", s = 10, label = 'Test data')
 
## plotting line for zero residual error
plt.hlines(y = 0, xmin = 0, xmax = 50, linewidth = 2)
 
## plotting legend
plt.legend(loc = 'upper right')
 
## plot title
plt.title("Residual errors")
 
## method call for showing the plot
plt.show()
```

```python
print("b0=")
b_0 = reg.intercept_
b_0
```

```python
#8
from sklearn.metrics import mean_squared_error
reg_fit = reg.fit(X, y)
y_pred = reg_fit.predict(X)
mse = mean_squared_error(y,y_pred)
mse
```
