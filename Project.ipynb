```python
df1_6_x = df1.copy()
df1_6_x.drop(["y"], axis = 1, inplace = True)
X = df1_6_x.to_numpy()
X
```

```python
df1_6_y = df1.copy()
df1_6_y.drop(['x1', 'x2', 'x3', 'x4', 'x5','x6','x7','x8','x9','x10','x11','x12','x13'], axis = 1, inplace = True)
df1_6_y

def flatten(t):
    return [item for sublist in t for item in sublist]

y = flatten(df1_6_y.to_numpy())
y
```

```python
#6-7
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(X, y)
b_i = reg.coef_
b_0 = reg.intercept_
print('Coefficients from b1 to b13: ', b_i)
print('Coefficient b0: ', b_0)
 
```

```python
#8
from sklearn.metrics import mean_squared_error
reg_fit = reg.fit(X, y)
y_pred = reg_fit.predict(X)
mse = mean_squared_error(y,y_pred)
mse
```
